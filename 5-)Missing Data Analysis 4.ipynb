# Methods of the filling  Missing Data 

```python
import pandas as pd
import numpy as np
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])

df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}        
)

df
```

```python
df["V1"].fillna(0)
```

```python
df["V1"].mean()
```

```python
df["V1"].fillna(df["V1"].mean())
```

```python
df.apply(lambda x:x.fillna(x.mean()),axis=0)#  in this code, axis =0 operates  on the basis of the  columns
```

```python
df.fillna(df.mean()[:])
```

```python
df.fillna(df.mean()["V1":"V2"])
# We assumed that the distribution of missing data  in the  v1 and v2 columns has  normal distirubition so we filled it with the mean
```

```python
df.fillna(df.median()["V3"])

#We assumed that the distribution of missing data  in the  v3 column has not   normal distirubition so we filled it with the median
```

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
V4 = np.array(["IT","IT","HR","HR","HR","HR","HR","IT","IT"])

df1 = pd.DataFrame(
        {"Salary" : V1,
         "V2" : V2,
         "V3" : V3,
        "departmant" : V4}        
)

df1
```

* Above the data set, filling the missing data in the SALARY column with a direct average may be incorrect.

* Because salaries may vary according to the departments.

 
* That's why we need to evaluate each department on its own and fill it with an average salary accordingly.


 
* That is, we get  the average of the salaries of the employees in the HR department.


* then  we fill out missing data in salaries HR employees by using  the average of the salaries of the employees in the HR department.



* if it is necessary  we can apply same process in IT departmant

```python
df1.groupby("departmant")["Salary"].mean()
```

```python
df1["Salary"].fillna(df1.groupby("departmant")["Salary"].transform("mean"))
```
