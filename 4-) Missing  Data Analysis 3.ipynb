```python
import missingno as msno
```

```python
import numpy as np
import pandas as pd

V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([12,np.NaN,5,6,14,np.NaN,7,2,31])

df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}        
)

df
```

```python
msno.bar(df);
```

* Above graph shows the total observations in the data set and the number of observations of missing values , the number of observations currently in the columns.

```python
df.isnull().sum()
```

```python
import seaborn as sns
sns.heatmap(df.isnull(),cbar=False);
```

* Whether the missing values is random or not can be determined by looking at the relationship between the variables.


* Visualization of missing values  is a good choice so as to understand whetherthe missing values is random or not 


* If we give an example, there  are generally  missing data  at the same index in the columns v2 and v3.


* this means that these two variable groups(v2&v3) are dependent.


* Thus there is no random missing values in the data set.

```python
msno.matrix(df);
```

```python
import seaborn as sns
df1 = sns.load_dataset("planets").copy()
df1.head()
```

```python
sns.heatmap(df1.isnull(),cbar=False);
```

* When we look at the heat map above, it is seen that "mass" and "distance" columns have missing data depending on each other.


* Because of the this reason  missing data is not randomly generated.

```python
msno.matrix(df1);
```

```python
msno.heatmap(df1);
```

* Above chart shows that how these variables affect each other in terms of missing values.


* It is seen that the variables here have missing values  depending on each other.


* The dependency of mass and distance regarding missing value  on this image is 0.5.

# Missing Values on Time Series Analysis

```python
null_pattern = (np.random.random(1000).reshape((50, 20)) > 0.5).astype(bool)
null_pattern
```

```python
null_pattern_df = pd.DataFrame(null_pattern).replace({False: None})

null_pattern_df
```

```python
msno.matrix(null_pattern_df.set_index(pd.period_range('1/1/2011', '2/1/2015', freq='M')) , freq='BQ');
```

```python
msno.heatmap(null_pattern_df);
```
