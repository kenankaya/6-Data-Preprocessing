# Variable Transformations Methods

**These kind of  transformations are made to perform mathematical operations on categorical variables in the data set.**

## 1-)  ZERO-ONE Transformation

* If there are only two class in a column(feature), we can use ZERO-ONE Transformation

```python
import seaborn as sns
tips = sns.load_dataset('tips')
df = tips.copy()
df_l = df.copy()
```

```python
df_l.head()
```

```python
df_l["new_sex"]=df_l["sex"].cat.codes
df_l.head()

# There are only two classes(male,female) in the "SEX" Column . thus we use method of ZERO-ONE Transformation 
```

* We transformed the **sex** column including categorical data into 0-1 values in a new column which is called **new_sex**


* **ONE** represents **FEMALE**


* **ZERO** represents **MALE**

```python
from sklearn import preprocessing

lbe = preprocessing.LabelEncoder()
```

```python
df_l["new_sex_1"] = lbe.fit_transform(df_l["sex"])
```

```python
df_l.head()
```

 0-1 TRANSFORMATION BY LABEL ENCODER IN SCKITLEARN
 
 PAY ATTENTION
 
 IT  HAS DIFFERENT REPRESANTATION.


* **ONE** represents **MALE**


* **ZERO** represents **FEMALE**

## 2-) Multi-Class Transformation

* If there are more than two classes in a column(feature), we can use Multi-Class Transformation

```python
lbe1= preprocessing.LabelEncoder()
```

```python
df_l["new_day"] = lbe.fit_transform(df_l["day"])

# There are MORE two classes(Monday,tuesday, wednesday....sunday) in the "DAY" Column . thus we use method of  Multi-Class Transformation 
```

```python
df_l.head(30)
```

* We perform Multi-Class transformations in columns(features) with more than 2 classes.



* After Multi-Class transformations is applied, the effects of the columns(features) in the new data set  on the dependent variable are different each other.



* Therefore it is necessary to make a tranformation which is  called ONE HOT ENCODEING





* On ONE HOT ENCODEING, there is a problem called the DUMMY VARIABLE TRAP



* To avoid the  DUMMY VARIABLE TRAP, we need to reduce the number of categorical variables(features) by one.

* That is, we need to extract the categorical variable(feature) that we have transformed from the new data set.


### 2.1-)One-Hot Transformation and Dummy Variable Trap

```python

df_one_hot = df.copy()
df_one_hot
```

### 2.1.1) One-hot encoding for "SEX" Column( feature)

```python
import pandas as pd 
 
df_one_hot=pd.get_dummies(df_one_hot, columns = ["sex"], prefix = ["sex"])
```

```python
df_one_hot.head()
```

Above chart ,two-class one-hot encoding transformation   has been done  


* **"Sex"** column has gone from the dataframe and **"Sex"** column transformed into **"sex_Male" & "sex_Female"** columns


* Two columns ,which are **"sex_Male" & "sex_Female"**, were added to Dataframe  

### 2.1.2) One-hot encoding for "DAY" Column( feature)

```python
df_one_hot=pd.get_dummies(df_one_hot, columns = ["day"], prefix = ["day"])
```

```python
df_one_hot.head()
```

Above chart ,Multi-class one-hot encoding transformation   has been done  


* **"Day"** column has gone from the dataframe and **"Day"** column transformed into **"day_Thur", "day_Fri","day_Sat","day_Sun"** columns


* Four columns ,which are **"day_Thur", "day_Fri","day_Sat","day_Sun"**, were added to Dataframe  

### 2.1.3) One-hot encoding for "TIME" Column( feature)

```python
df_one_hot.head()
```
